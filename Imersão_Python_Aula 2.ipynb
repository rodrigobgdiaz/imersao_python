```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Imersão Python - Tabela de ações_pura.xlsx", sheet_name="Principal")
df_principal.head(3)
```

```python
df_total_acoes = pd.read_excel("/content/Imersão Python - Tabela de ações_pura.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head(3)
```

```python
df_ticker = pd.read_excel("/content/Imersão Python - Tabela de ações_pura.xlsx", sheet_name="Ticker")
df_ticker.head(3)
```

```python
df_chatgpt = pd.read_excel("/content/Imersão Python - Tabela de ações_pura.xlsx", sheet_name="ChatGPT")
df_chatgpt.head(3)
```

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal.head(3)
```

```python
df_principal = df_principal.rename(columns={'Ativo':'ativo', 'Data':'data', 'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
df_principal.head(3)
```

```python
df_principal['var_pct'] = df_principal['var_dia_pct']/100
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal.head(3)
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='ativo', right_on='Código', how='left')
df_principal.head(3)
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal.head(3)
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal.head(3)
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal.head(3)
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal.head(3)
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal.head(3)
```

```python
df_principal['Resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável')).copy()
df_principal.head(3)
```

```python
df_principal = df_principal.merge(df_ticker, left_on='ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal.head(3)
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal.head(3)
```

```python
df_principal = df_principal.rename(columns={'Idade em anos':'Idade'}).copy()
df_principal.head(3)
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100')).copy()
df_principal.head(3)
```

```python
# Calculando o maior valor
maior = df_principal['variacao_rs'].max()
#Calculando o menor valor
menor = df_principal['variacao_rs'].min()
# Calculando a media
media = df_principal['variacao_rs'].mean()
# Calculando a media de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['variacao_rs'].mean()
# Calculando a media de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao_rs'].mean()

# Imprimir os resultados
print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Media\tR$ {media:,.2f}")
print(f"Media de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Media de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
# Novo dataframe somente para os resultados positivos
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu.head(3)
```

```python
# Novo dataframe somente para analise dos segmentos
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
# Grafico de pizza
fig_seg = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento')
fig_seg.show()
```

```python
# Novo dataframe somente para analise dos saldos
df_analise_saldo = df_principal.groupby('Resultado')['variacao_rs'].sum().reset_index()
# Formatando os numeros
pd.options.display.float_format = '{:,.2f}'.format
df_analise_saldo
```

```python
# Grafico de barras
fig_bars = px.bar(df_analise_saldo, x = 'Resultado', y = 'variacao_rs', text = 'variacao_rs', title='Variação Reais por Resultado')
fig_bars.show()
```

```python
# Novo dataframe somente para analise dos idade
df_analise_idade = df_principal.groupby('Cat_idade')['variacao_rs'].sum().reset_index()
df_analise_idade
```

```python
# Grafico de barras por idade
fig_idade = px.bar(df_analise_idade, x = 'Cat_idade', y='variacao_rs', text_auto = ',.2f', title='Variação Reais por Categoria de Idade')
fig_idade.update_traces(textposition="outside")
fig_idade.show()
```
